```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
file_path1 = "../5_data_enrichment/my_addresses_geocoded.csv"
file_path2 = "../4_integration/integration_definitive.xlsx"

not_found_trip = pd.read_csv(file_path1)
integration_definitive = pd.read_excel(file_path2)
```

```python
# Create a new column with unique labels
#integration_definitive['new_index'] = range(len(integration_definitive))

# Set place id as the index for integrated data
integration_definitive = integration_definitive.set_index('place_id')

# Set trip_id as index for not found trip data
not_found_trip = not_found_trip.set_index('id_trip')
```

```python
# remove ? char and put nan instead
integration_definitive= integration_definitive.replace('?', np.nan)
not_found_trip = not_found_trip.replace('?', np.nan)
#convert to float
integration_definitive['price_level_trip'] = integration_definitive['price_level_trip'].astype(float)
integration_definitive['price_level_g'] = integration_definitive['price_level_g'].astype(float)
```

```python
# Calculate rating weighting the trip rating and google rating on the total number reviews
has_rating_trip = ~integration_definitive['rating_trip'].isnull()
has_rating_g = ~integration_definitive['rating_g'].isnull()
has_both_ratings = has_rating_trip & has_rating_g

integration_definitive.loc[has_both_ratings, 'rating'] = ((integration_definitive.loc[has_both_ratings, 'rating_trip'] * integration_definitive.loc[has_both_ratings, 'total_reviews_trip'])
                                        + (integration_definitive.loc[has_both_ratings, 'rating_g'] * integration_definitive.loc[has_both_ratings, 'total_reviews_g'])) / (integration_definitive.loc[has_both_ratings, 'total_reviews_trip'] + integration_definitive.loc[has_both_ratings, 'total_reviews_g'])

# Calculate price_level weighting on total reviews
has_price_level_trip = ~integration_definitive['price_level_trip'].isnull()
has_price_level_g = ~integration_definitive['price_level_g'].isnull()

integration_definitive.loc[has_price_level_trip & has_price_level_g, 'price_level'] = ((integration_definitive.loc[has_price_level_trip & has_price_level_g, 'price_level_trip'] * integration_definitive.loc[has_price_level_trip & has_price_level_g, 'total_reviews_trip'])
                                                  + (integration_definitive.loc[has_price_level_trip & has_price_level_g, 'price_level_g'] *
                                                     integration_definitive.loc[has_price_level_trip & has_price_level_g, 'total_reviews_g'])) / (integration_definitive.loc[has_price_level_trip & has_price_level_g, 'total_reviews_trip'] + 
                                                                                                                               integration_definitive.loc[has_price_level_trip & has_price_level_g, 'total_reviews_g'])

integration_definitive.loc[has_price_level_trip & ~has_price_level_g, 'price_level'] = integration_definitive.loc[has_price_level_trip & ~has_price_level_g, 'price_level_trip']

integration_definitive.loc[~has_price_level_trip & has_price_level_g, 'price_level'] = integration_definitive.loc[~has_price_level_trip & has_price_level_g, 'price_level_g']

# Calculate total_reviews as sum of trip and google total reviews
integration_definitive['total_reviews'] = integration_definitive['total_reviews_g'].fillna(0) + integration_definitive['total_reviews_trip'].fillna(0)
integration_definitive
```

```python
import ast
for index, row in integration_definitive.iterrows():
    # Parse the 'geometry' field as a dictionary
    geometry_dict = ast.literal_eval(row['geometry'])
    
    # Extract the 'lat' and 'lng' values
    lat_value = geometry_dict['lat']
    lng_value = geometry_dict['lng']
    
    # Round the values to the third decimal place
    lat_rounded = round(lat_value, 10)
    lng_rounded = round(lng_value, 10)
    
    # Assign the rounded values to new columns 'latitude' and 'longitude'
    integration_definitive.at[index, 'latitude'] = lat_rounded
    integration_definitive.at[index, 'longitude'] = lng_rounded
```

```python

#take only useful fields. In order to avoid redaundancy we choose to keep just name and address of trip advisor
integration_definitive = integration_definitive[['name_trip','cook_type_trip','address_trip','latitude','longitude','rating','price_level','total_reviews','reviews']]
integration_definitive = integration_definitive.rename(columns={'name_trip': 'name', 'cook_type_trip': 'cook_type', 'address_trip': 'address'})
not_found_trip = not_found_trip[['name_trip','cook_type_trip','address_trip','latitude','longitude','rating_trip','price_level_trip','total_reviews_trip','reviews']]

not_found_trip = not_found_trip.rename(columns={'id_trip': 'place_id','name_trip': 'name', 'cook_type_trip': 'cook_type', 'address_trip': 'address', 'rating_trip': 'rating', 'price_level_trip': 'price_level','total_reviews_trip': 'total_reviews'})

integration_definitive = pd.concat([integration_definitive, not_found_trip])

```
