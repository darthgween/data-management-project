```python
# Create a new column with unique labels
df2['new_index'] = range(len(df2))

# Set the new column as the index
df2 = df2.set_index('new_index')
```

```python
df2= df2.replace('?', np.nan)
```

```python
# # Calculate weighted average rating with handling missing or zero values
# df2['rating'] = ((df2['rating_trip'] * df2['total_reviews_trip']) + (df2['rating_g'] * df2['total_reviews_g'])) / (df2['total_reviews_trip'] + df2['total_reviews_g'])
df2['price_level_trip'] = df2['price_level_trip'].astype(float)
df2['price_level_g'] = df2['price_level_g'].astype(float)
# #df2['price_level_g'] = df2['price_level_g'].replace('?', np.nan)
# df2['price_level'] = ((df2['price_level_trip'] * df2['total_reviews_trip']) + (df2['price_level_g'] * df2['total_reviews_g'])) / (df2['total_reviews_trip'] + df2['total_reviews_g'])
# df2['total_reviews']=df2['total_reviews_g']+df2['total_reviews_trip']
```

```python
# Calculate rating
has_rating_trip = ~df2['rating_trip'].isnull()
has_rating_g = ~df2['rating_g'].isnull()
has_both_ratings = has_rating_trip & has_rating_g

df2.loc[has_both_ratings, 'rating'] = ((df2.loc[has_both_ratings, 'rating_trip'] * df2.loc[has_both_ratings, 'total_reviews_trip'])
                                        + (df2.loc[has_both_ratings, 'rating_g'] * df2.loc[has_both_ratings, 'total_reviews_g'])) / (df2.loc[has_both_ratings, 'total_reviews_trip'] + df2.loc[has_both_ratings, 'total_reviews_g'])

# Calculate price_level
has_price_level_trip = ~df2['price_level_trip'].isnull()
has_price_level_g = ~df2['price_level_g'].isnull()

df2.loc[has_price_level_trip & has_price_level_g, 'price_level'] = ((df2.loc[has_price_level_trip & has_price_level_g, 'price_level_trip'] * df2.loc[has_price_level_trip & has_price_level_g, 'total_reviews_trip'])
                                                  + (df2.loc[has_price_level_trip & has_price_level_g, 'price_level_g'] *
                                                     df2.loc[has_price_level_trip & has_price_level_g, 'total_reviews_g'])) / (df2.loc[has_price_level_trip & has_price_level_g, 'total_reviews_trip'] + 
                                                                                                                               df2.loc[has_price_level_trip & has_price_level_g, 'total_reviews_g'])

df2.loc[has_price_level_trip & ~has_price_level_g, 'price_level'] = df2.loc[has_price_level_trip & ~has_price_level_g, 'price_level_trip']

df2.loc[~has_price_level_trip & has_price_level_g, 'price_level'] = df2.loc[~has_price_level_trip & has_price_level_g, 'price_level_g']

# Calculate total_reviews
df2['total_reviews'] = df2['total_reviews_g'].fillna(0) + df2['total_reviews_trip'].fillna(0)
```

```python

fig, ax = plt.subplots(figsize=(60, 60))
sns.boxplot(x='cook_type_trip', y='rating', data=df_integrated, ax=ax, showfliers=False, palette='Set2')
ax.set_xticklabels(ax.get_xticklabels(), rotation=90, ha='right', fontsize=7)
ax.set_xticklabels([x.get_text().replace(', ', '\n') for x in ax.get_xticklabels()])
plt.subplots_adjust(bottom=0.4) # Adjust the bottom margin to make room for the labels
plt.title('Box Plot of Ratings by Cuisine Type', fontsize=16)
plt.xlabel('Cuisine Type', fontsize=12)
plt.ylabel('Rating', fontsize=12)
plt.show
```

```python
# Calculate the correlation matrix
corr_matrix = df_integrated[['total_reviews', 'price_level', 'rating']].corr()

# Create a heatmap of the correlation matrix
fig, ax = plt.subplots(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', ax=ax)
ax.set_title('Correlation Matrix of Trip Advisor and Google Reviews', fontsize=16)
plt.show()

# Create a scatterplot of price level vs rating
fig, ax = plt.subplots(figsize=(8, 6))
sns.scatterplot(x='price_level', y='rating', data=df_integrated, ax=ax, hue='cook_type_trip', alpha=0.7, s=50)
ax.set_title('Scatterplot of Price Level vs Rating', fontsize=16)
ax.set_xlabel('Price Level', fontsize=12)
ax.set_ylabel('Rating', fontsize=12)
plt.show()
```
