```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from geopy.geocoders import Nominatim
import time
```

```python
file_path1 = "./my_addresses_geocoded.csv"

not_found = pd.read_csv(file_path1)
```

```python
#Get the coordinates of the non matched trip advisor restaurants

# create a Nominatim geocoder object
geolocator = Nominatim(user_agent='myapplication')


# create new columns for latitude and longitude
# not_found['latitude'] = ''
# not_found['longitude'] = ''

# loop over each row in the DataFrame and geocode the address
#for i, row in not_found.iterrows():
for i, row in not_found.iloc[4000:].iterrows():
    address = row['address_trip']
    location = geolocator.geocode(address) # get the location object for the address
    if location:
        not_found.at[i, 'latitude'] = location.latitude # update the latitude for this row
        not_found.at[i, 'longitude'] = location.longitude # update the longitude for this row
        print(f"Geocoded address {i+1}: {address} -> ({location.latitude}, {location.longitude})")
    else:
        print(f"Failed to geocode address {i+1}: {address}")
    time.sleep(1) # wait for 1 second before making the next request

    # save the updated DataFrame to a new CSV file every 10 records
    if (i+1) % 10 == 0:
        filename = f"my_addresses_geocoded.csv"
        not_found.to_csv(filename, index=False)
        print(f"Saved geocoded addresses to {filename}")

# save the final version of the DataFrame to a new CSV file
not_found.to_csv('my_addresses_geocoded.csv', index=False)

```
