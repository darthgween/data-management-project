```python
import re

# Regular expression pattern to match the relevant information
pattern = r"\{'score': \d+, 'google': '([^']*)', 'formatted_google': '([^']*)'\}"

# Function to extract information from the "similar" column
def extract_similar(similar):
    matches = re.findall(pattern, similar)
    return {match[0]: match[1] for match in matches}

# Apply the function to the "similar" column
merged_df['similar'] = merged_df['similar'].apply(extract_similar)
```

```python
import pandas as pd
from concurrent.futures import ThreadPoolExecutor
from fuzzywuzzy import fuzz



# preprocess the address columns in both datasets
merged['address_clean'] = merged['address_trip'].str.replace(r'\b\d{5}\b', '').str.lower().str.strip()
google_dataset['address_clean'] = google_dataset['address'].str.replace(r'\b\d{5}\b', '').str.lower().str.strip()

# define a function to compute the similarity score between two strings
def compute_similarity(str1, str2):
    return fuzz.token_set_ratio(str1, str2)

# define a function to match a row from merged with a row from google_dataset
def match(row1):
    title1 = row1['title']
    address1 = row1['address_clean']
    scores = []
    for _, row2 in google_dataset.iterrows():
        title2 = row2['new_column']
        address2 = row2['address_clean']
        score1 = compute_similarity(address1, address2)
        score2 = compute_similarity(title1, title2)
        final_score = fuzz.WRatio(address1 + title1, address2 + title2)
        if final_score > 80:
            scores.append((final_score, row1.name, row2.name))
    if len(scores) > 0:
        scores.sort(reverse=True)
        return scores[0]

# match the rows in parallel using ThreadPoolExecutor
results = []
with ThreadPoolExecutor() as executor:
    futures = [executor.submit(match, row) for _, row in merged.iterrows()]
    for future in futures:
        res = future.result()
        if res:
            results.append(res)

# create a new DataFrame with the matched rows
matched = pd.DataFrame(results, columns=['score', 'merged_index', 'google_index'])
matched.set_index('merged_index', inplace=True)
matched.sort_index(inplace=True)

# merge the two datasets using the matched rows
final = merged.merge(google_dataset, left_index=True, right_index=True, suffixes=('_merged', '_google'), how='outer')
final.set_index('title', inplace=True)

for _, row in matched.iterrows():
    merged_row = merged.loc[row.name]
    google_row = google_dataset.loc[row['google_index']]
    final_row = final.loc[merged_row['title']]
    final_row['score'] = row['score']
    final_row['address_google'] = google_row['address']
    final_row['price_level_google'] = google_row['price_level_google']
    final_row['rating_google'] = google_row['rating_google']
    final_row['total_reviews_google'] = google_row['total_reviews_google']

final.to_csv('matched.csv')
```
