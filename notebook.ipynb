droppiamo i duplicati. sostituiamo il numerino con uno spazioe sostistuiamo il simbolo degli euro con un numero
nb si stiamo sosituendo il numerino all'inizio basta, nella regex ^\d+\. in questo modo se abbiamo un ristorante che ha il numerino
e oltre a quello inizia con un numero nel nome (ex: 100 montadidos) noi avremo il nome corretto con il numerino

```python

# Syntax of drop_duplicates
df = df.drop_duplicates(subset='title')
df['title'] = df['title'].str.replace('^\d+\.', '')
df['expensive'] = df['expensive'].replace({'€': 1, '€€': 2,'€ - €€': 1.5,'€€€': 3, '€€ - €€€': 2.5, '€€€€':4 , '€€€ - €€€€':3.5, '€€€€ - €€€€€':4.5,'€€€€€':5 })

df
```
