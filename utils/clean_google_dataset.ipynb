Step 1. merge permanentrly_closed column (boolean) into business_status (string)

```python
df.loc[df['permanently_closed'] == True, 'business_status'] = 'PERMANENTLY_CLOSED' 
df = df.drop(columns=['permanently_closed'])
df
```

Step 2. Adjust Geometry field according to what we need

```python
df['geometry'] = df['geometry'].apply(lambda x: ast.literal_eval(x)['location'])
df['geometry']
```

```python
df = df.drop(columns=['icon','icon_background_color','icon_background_color','icon_mask_base_uri', 'opening_hours', 'photos', 'plus_code', 'reference', 'scope'])
df
```
