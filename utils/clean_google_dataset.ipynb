Step 1. merge permanentrly_closed column (boolean) into business_status (string)

```python
df.loc[df['permanently_closed'] == True, 'business_status'] = 'PERMANENTLY_CLOSED' 
df = df.drop(columns=['permanently_closed'])
```

```python
df = df[df['business_status'] != 'PERMANENTLY_CLOSED']
count = (df['address_g'].str.contains(', Milano|italy')).sum()
count
```

Step 2. Adjust Geometry field according to what we need

```python
df['geometry'] = df['geometry'].apply(lambda x: ast.literal_eval(x)['location'])
df['geometry']
```

```python
df = df.drop(columns=['icon','icon_background_color','icon_background_color','icon_mask_base_uri', 'opening_hours', 'photos', 'plus_code', 'reference', 'scope'])
df
```

```python
df.isna().sum()
```

```python
# Check if there are some restaurants with the same name
df_grouped = df.groupby('name').size().reset_index(name='counts')
result = df_grouped[df_grouped['counts'] > 1]
result.to_csv('../output/duplicates_name.csv')
```

```python
new_col_names = {'price_level': 'price_level_g',
                'rating': 'rating_g',
                'user_ratings_total': 'total_reviews_g',
                'price_level': 'price_level_g',
                'vicinity': 'address_g',
                'name': 'name_g'}
df = df.rename(columns=new_col_names)
df
```
